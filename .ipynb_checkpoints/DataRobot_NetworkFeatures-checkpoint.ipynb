# Create Date Partition For Train, Validation, Holdout

```python
dtrng_h = pd.date_range('2015-05-01', '2016-07-30', freq='M').format(formatter=lambda x: x.strftime('%Y-%m-%d'))
dtrng_v = pd.date_range('2015-03-01', '2016-05-30', freq='M').format(formatter=lambda x: x.strftime('%Y-%m-%d'))
```

```python
sliding_window_dates = list(zip(dtrng_v, dtrng_h))
sliding_window_dates
```

```python
def custom_partition(dt, v, h):
    if dt >= pd.to_datetime(h):
        return 'h'
    elif ((dt >= pd.to_datetime(v)) & (dt < pd.to_datetime(h))):
        return 'v'
    else:
        return 't'
```

# Send To DataRobot

```python
import datarobot as dr
from datarobot import UserTVH
#Create custom partition object
# part = UserTVH(user_partition_col = 'part', training_level = 't', 
#                validation_level ='v', holdout_level ='h', seed=0)
```

```python
droplist = ['missed_pmts', 'nonfraud_default', 'sift_labeled_bad', 'sift_science_score', 
            'made_pmts', 'fraud_flag', 'reserve_fee_pct','country_id', 'created_at', 
            'disbursed_at', 'country_internet_users', 'country_life_expectancy',
            'country_literacy','country_gdppc', 'country_id', 'name']

def send_to_datarobot(df, proj_name, cols_to_drop, val_date, holdout_date, metric=None):
    """
    helper function that takes data and uploads to DataRobot
    """
    #Filter and copy; throw away all data one month after holdout as holdout only needs to be one month
    mask = df.created_at < (pd.to_datetime(holdout_date) + np.timedelta64(1, 'M'))
    df = df.copy()[mask]
    
    #apply partition logic
    df['part'] = df.created_at.apply(lambda x: custom_partition(x, val_date, holdout_date))
    part = UserTVH(user_partition_col = 'part', training_level = 't', 
               validation_level ='v', holdout_level ='h', seed=0)
    print(df.groupby('part')['created_at'].agg(['min', 'max', 'count']))
    
    #drop irrelevant columns
    cols_to_drop = [x for x in cols_to_drop if x in df.columns]
    final_df = df.drop(cols_to_drop, axis=1)
    #Combine the ids to obtain a unique user id
    final_df['unique_id'] = final_df.borrower_id.astype(str) +'__'+ final_df.id.astype(str)
    final_df.drop(['borrower_id', 'id'], axis = 1, inplace=True)
    
    #save file to disk, with compression
    file_path = '/Users/hamelhusain/Google Drive/Team Shared Folder - Zidisha/DR_'+proj_name+'.gz'
    print('Saving {}'.format(file_path))
    final_df.to_csv(file_path, compression = 'gzip', index=False)
    
    print('Beginning DR Upload')
    proj = dr.Project.create(file_path, project_name=proj_name)
    proj.set_target(target = 'default_flag', 
                partitioning_method = part,
                worker_count = 5,
                metric=metric)
    return proj
```

```python
for val, hold in sliding_window_dates:
    proj_name = 'zid_sw_' + hold
    send_to_datarobot(df, proj_name, droplist, val, hold)
```

# Retrieve Leaderboard

```python
from pandas.io.json import json_normalize

cols_to_exclude = ['metrics.LogLoss.crossValidation', 'metrics.LogLoss.validation']

def get_leaderboard(proj, cols_to_exclude=cols_to_exclude):
    """get leader board from project object, which you can get from dr.Projectlist()"""
    proj.unlock_holdout()
    leaderboard = json_normalize([x.__dict__ for x in proj.get_models()])
    leaderboard['Project_Name'] = proj.project_name
    leaderboard['Holdout_Date'] = proj.project_name[-10:]
    cols = [col for col in leaderboard.columns if col[:1] != '_' and not ('metric' in col and proj.metric not in col) and \
            col not in cols_to_exclude]
    return leaderboard[cols]
```

```python
leaderboards = []
for project in Projects:
    leaderboards.append(get_leaderboard(project))
```

```python
all_lb = pd.concat(leaderboards)
all_lb.to_csv('/Users/hamelhusain/Google Drive/Team Shared Folder - Zidisha/TimeSeriesCV_Results.csv', index=False)
```
